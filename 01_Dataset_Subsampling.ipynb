# Dataset Subsampling

```python
import pandas as pd
from tqdm import tqdm
```

```python
reviews_dfs = [
    pd.read_csv('reviews-1-115.csv'),
    pd.read_csv('reviews-115-1230.csv'),
    pd.read_csv('reviews-1230-2345.csv'),
    pd.read_csv('reviews-2345-4575.csv'),
    pd.read_csv('reviews-4575-6805.csv'),
    pd.read_csv('reviews-6805-9035.csv'),
    pd.read_csv('reviews-9035-11265.csv'),
    pd.read_csv('reviews-11265-13495.csv'),
    pd.read_csv('reviews-13495-13500.csv'),
    pd.read_csv('reviews-13500-13537.csv'),
    pd.read_csv('reviews-13537-27075.csv')
]
```

```python
# Concatenate the DataFrames containing portions of the dataset
reviews_df = reviews_dfs.pop(0)
for x in tqdm(range(len(reviews_dfs))):
    reviews_df = pd.concat([reviews_df, reviews_dfs[0]])

    # Delete the portion of data that is no longer needed
    del reviews_dfs[0]

# Reset the index to keep it aligned with the actual rows
reviews_df.reset_index(drop=True, inplace=True)
```

```python
reviews_df
```

```python
reviews_df.to_csv('reviews.csv', index=False)
```

```python
# Randomly sample 2% of the whole dataset to carry out the initial analysis
small_reviews_df = reviews_df.sample(frac=0.02, random_state=0)
```
